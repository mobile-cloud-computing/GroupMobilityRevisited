# Mapping All Datapoints from GPS

```python
df_Gps =  pd.read_csv("Datasets/gps.csv")
df_Gps.drop(df_Gps.filter(regex="Unnamed"), axis=1, inplace=True)
```

```python
df_Gps['coordinates'] = list(zip(df_Gps['longitude'],df_Gps['latitude']))
df_Gps['coordinates'] = df_Gps['coordinates'].apply(Point)
df_Gps = gpd.GeoDataFrame(df_Gps, geometry='coordinates')
df_Gps
```

### Verifying the GPS Coordinates

```python
df_Gps.longitude.round(6).max(),df_Gps.round(6).longitude.min()
```

```python
df_Gps.latitude.round(6).max(),df_Gps.round(6).latitude.min()
```

```python
fig,ax = plt.subplots(figsize = (15,15))
shanghai.boundary.plot(ax=ax, alpha = 0.4, color="black")
df_Gps.plot(ax=ax, markersize = 15,color = "purple", marker ="o", legend=True)
```

## Finding recods count on Time Bases

```python
df_verifiedStations_AppUsage['time']= pd.to_datetime(df_verifiedStations_AppUsage['time'])
df_verifiedStations_AppUsage['hour'] = df_verifiedStations_AppUsage['time'].dt.hour
```

```python
# 0 => 2017-08-21 00:00:00  --  2017-08-21 00:59:59	
# 1 => 2017-08-21 01:00:00  --  2017-08-21 01:59:59
df_specificTime = pd.DataFrame()
def findAppUsageforSpecificTimeHour(hour): 	
    df_specificTime = df_verifiedStations_AppUsage[df_verifiedStations_AppUsage.hour == hour]
    return df_specificTime 
```
